Get page content from this wiki

```python
import requests
newsurl = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
res = requests.get(newsurl)
res.encoding = 'utf-8'
print(res.text)
```

Use BeautifulSop to parse the table of Postcode, Borough and Neighbourhood
from the response html.

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'lxml')
table=soup.find_all('table')[0]
table
```

use panda.read_html method to read the table data to a panda DataFrame.

```python
df=pd.read_html(str(table),header=0)[0]
df
```

Drop the lines with a borough that is Not assigned.

```python
df1 = df[~df['Borough'].isin(['Not assigned'])]
df1.head()
```

Merge the lines that have the same Postcode, combine the Neighbourhoods in one row

```python
f=lambda x : ', '.join(x)

df2=pd.DataFrame(df1.groupby(['Postcode','Borough'])['Neighbourhood'].apply(f)).reset_index()
df2
```

Set the Neighbourhood that is Not assigned to Borough, which is Queen's Park, the only one row.

```python
df2['Neighbourhood'].replace(['Not assigned'],'Queen\'s Park', inplace=True)
```
